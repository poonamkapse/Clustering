# Kmeans Clustering

```python
from sklearn.cluster import KMeans
from scipy.spatial.distance import cdist
```

```python
#Normalizing  data fuction
def norm_func(i):
    x=(i-i.min())/(i.max()-i.min())
    return(x)
```

```python
df_normal =norm_func(data.iloc[:,1:])
df_normal
```

```python
# Elbow curve
# distortion
wcss=[]
for i in range(1,11):
    kmeans=KMeans(n_clusters=i)
    kmeans.fit(df_normal)
    wcss.append(kmeans.inertia_)
plt.plot(range(1,11),wcss)
plt.title('Elbow curve')
plt.xlabel('Number of clusters')
plt.ylabel('WCSS')
plt.show()
```

```python
# from above graph k=4 is the optimal value for k
model=KMeans(n_clusters=4)
model.fit(df_normal)
model.labels_
```

```python
x=pd.Series(model.labels_)
data['Clust']=x
```

# DBSCAN clustering

```python
from sklearn.cluster import DBSCAN
from sklearn.preprocessing import StandardScaler
```

```python
data1=data.iloc[:,1:5]
```

```python
data1.values
```

```python
stscaler=StandardScaler().fit(data1.values)
x=stscaler.transform(data1.values)
x
```

```python
dbscan=DBSCAN(eps=2,min_samples=5)
dbscan.fit(x)
```

```python
dbscan.labels_
```

```python
cluster=pd.DataFrame(dbscan.labels_,columns=['cluster'])
```

```python
pd.concat([data1,cluster],axis=1)
```

# Heirarchical Clustering

```python
data2 =pd.read_csv('crime_data.csv')
data2.head()
```

```python
# Normalization function 
def norm_func(i):
    x = (i-i.min())/(i.max()-i.min())
    return x
```

```python
# Normalized data frame here we are considering the numerical part of data
df_norm = norm_func(data2.iloc[:,1:])
```

```python
df_norm.iloc[:,:]
```

```python
# creating dendrogram
dendrogram = sch.dendrogram(sch.linkage(df_norm, method='complete'))
plt.figure(figsize=(10,7))
     
```

```python
#AgglomerativeClustering
# creatinng  clusters
from sklearn.cluster import AgglomerativeClustering
cluster2 = AgglomerativeClustering(n_clusters=3, affinity = 'euclidean', linkage = 'complete')
```

```python
cluster2.fit_predict(df_norm)
```

```python
data2['clusterid']=cluster2.labels_
data2
```
